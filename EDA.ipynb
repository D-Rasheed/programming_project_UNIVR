```python
#General libraries
import pandas as pd
import numpy as np
#Visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
#Importing data
eda=pd.read_csv('C:/Users/danis/Downloads/archive-Telco customer churn/Clean_test3.csv')
```

```python
eda.Churn.value_counts()
```

```python
categorical_features = []
numerical_features = []

for col in eda.columns:
  if len(eda[col].unique())<=10:
    categorical_features.append(col)
  else:
    numerical_features.append(col)

print('numerical_features = ', numerical_features)
print('categorical_features = ', categorical_features)
```

```python
# Histribution of numerical features

plt.figure(figsize=(15,4))
plt.subplot(131)
sns.histplot(eda['tenure'])
plt.subplot(132)
sns.histplot(eda['MonthlyCharges'])
plt.subplot(133)
sns.histplot(eda['TotalCharges'])
```

```python
# Countplot of some important categorical features

plt.figure(figsize=(20,5))
plt.subplot(151)
sns.countplot(eda, x='Churn', hue='gender');
plt.subplot(152)
sns.countplot(eda, x='Churn', hue='SeniorCitizen');
plt.subplot(153)
sns.countplot(eda, x='Churn', hue='Partner');
plt.subplot(154)
sns.countplot(eda, x='Churn', hue='Dependents');
plt.subplot(155)
sns.countplot(eda, x='Churn', hue='PhoneService');
```

```python
# Boxplot to investigate the numerical features variations against the target column(churn)

plt.figure(figsize=(15,4))
plt.subplot(131)
sns.boxplot(eda, x='Churn', y='tenure');
plt.subplot(132)
sns.boxplot(eda, x='Churn', y='MonthlyCharges');
plt.subplot(133)
sns.boxplot(eda, x='Churn', y='TotalCharges');
```

```python
# Scatterplots of numerical features with hue = 'Churn' which is our target

plt.figure(figsize=(15,4))
plt.subplot(131)
sns.scatterplot(x=eda['MonthlyCharges'], y=eda['TotalCharges'], hue=eda['Churn']);
plt.subplot(132)
sns.scatterplot(x=eda['tenure'], y=eda['MonthlyCharges'], hue=eda['Churn']);
plt.subplot(133)
sns.scatterplot(x=eda['tenure'], y=eda['TotalCharges'], hue=eda['Churn']);
```
